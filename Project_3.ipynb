```python
#Imports

import requests
import pandas as pd
from pandas.io.json import json_normalize

```

```python
#Sample API request, will need to be dynamic

response = requests.get('https://sandbox.tradier.com/v1/markets/history',
    params={'symbol': 'MSFT', 'interval': 'daily', 'start': '2000-01-01', 'end': '2020-03-06'},
    headers={'Authorization': 'Bearer IGef006CVwMc97oonnuCq28ce1dK', 'Accept': 'application/json'}
)
json_response = response.json()

json_response
```

```python
#Basic data collected, converted to datetime and weekday parsed

df = pd.DataFrame.from_dict(json_normalize(json_response["history"]['day']), orient='columns')

df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')

df['Weekday'] = df['date'].dt.day_name()

df
```
